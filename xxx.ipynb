```python
import json
import warnings

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
pd.set_option('display.max_rows', 100, 'display.max_columns', 1000, "display.max_colwidth", 1000, 'display.width', 1000)

data = pd.read_excel("./data/Second-Hand House.xlsx", na_values=np.nan)

# 看一下数据维度
print(data.shape)

data = data[data.columns[:-5]]

data['建筑面积'] = data['建筑面积'].str[:-1]
data['建筑面积'] = data['建筑面积'].astype(float)
data.head()
```

```python
data['地段'] = data['地段'].str[2:-2]
data['地段'] = data['地段'].str.replace('\'', '')
data['地段'].value_counts()
```

```python
data['地段_区'] = data['地段'].str.split(',', expand=True)[0]
data['地段_区'].value_counts()
```

```python
data['地段_路'] = data['地段'].str.split(',', expand=True)[1]
data['地段_路'].value_counts()
```

```python
data['房屋户型_室'] = data['房屋户型'].str[0]
data['房屋户型_厅'] = data['房屋户型'].str[2]
data['房屋户型_厨'] = data['房屋户型'].str[4]
data['房屋户型_卫'] = data['房屋户型'].str[6]
data.head()
```

```python
data['楼层位置'] = data['所在楼层'].str.split('(', expand=True)[0]
data['楼层位置'].value_counts()
```

```python
data['楼层数'] = data['所在楼层'].str.split('(', expand=True)[1]
data['楼层数'] = data['楼层数'].str[1:-2].astype(int)
data['楼层数'].value_counts()
```

```python
count = 0
areaDiff = 0
for i in range(data.shape[0]):
    if "㎡" in data.iloc[i]['套内面积']:
        areaDiff += data.iloc[i]['建筑面积'] - float(data.iloc[i]['套内面积'][:-1])
        count += 1
print(round(areaDiff,2), count, round(areaDiff/count, 2))
```

```python
# 定义填充函数
def fillInsideArea(df):
    if "㎡" in df['套内面积']:
        return float(df['套内面积'][:-1])
    else:
        return (df['建筑面积'] - 21.03)

# 填充
data['套内面积'] = data.apply(lambda x:fillInsideArea(x), axis=1)
data.head()
```

```python
data['房屋朝向'].replace(' ', '', regex=True, inplace=True)
data['房屋朝向'].value_counts()
```

```python
data['所在楼层'].value_counts()
```

```python
data['挂牌时间'] = data['挂牌时间'].apply(str)
data = data[-data['挂牌时间'].str.contains('^4|00:00:00')]
data['挂牌时间'] = pd.to_datetime(data['挂牌时间'])
data['挂牌时间'].value_counts()
```

```python
data['上次交易'] = data['上次交易'].apply(str)
data = data[-data['上次交易'].str.contains('^4|00:00:00')]
data['上次交易'] = pd.to_datetime(data['上次交易'])
data['上次交易'].value_counts()
```

```python
data['时差'] = (data['挂牌时间'] - data['上次交易']).dt.total_seconds().astype(int)
data['梯户比例'].head()
```

```python
data['楼梯数'] = data['梯户比例'].str.split('梯', expand=True)[0]
data['楼梯数'].head()
```

```python
data['楼梯数'] = data['楼梯数'].map({'一': 1, '二': 2, '两': 2, '三': 3, '四': 4, '五': 5, '六': 6, '七': 7, '八': 8, '九': 9})
data['楼梯数'].head()
```

```python
data['住户数'] = data['梯户比例'].str.split('梯', expand=True)[1]
data['住户数'] = data['住户数'].str[:-1]
data['住户数'].value_counts()
```

```python
data['住户数'] = data['住户数'].map({'一': 1, '二': 2, '两': 2, '三': 3, '四': 4, '五': 5, '六': 6, '七': 7, '八': 8, '九': 9, '十一': 12, '十二': 12, '十三': 12, '十四': 12, '十五': 12, '十六': 12, '十七': 12, '十八': 12, '十九': 12, '二十': 16, '二十一': 12, '二十二': 12, '二十三': 12, '二十四': 24, '二十五': 12, '二十六': 12, '二十八': 12, '二十九': 12, })
data['住户数'].value_counts()
```

```python
data['梯户比例'] = data.apply(lambda df: df['楼梯数']/df['住户数'], axis=1)
data['梯户比例'].value_counts()
```

```python
data.to_excel("./data/Second-Hand House plus.xlsx")
```
